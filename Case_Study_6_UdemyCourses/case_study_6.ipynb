# 8. Find out number of courses per subjects

```python
df['subject'].value_counts()
```

```python
sns.countplot(x='subject', data=df)
plt.xlabel('Subjects')
plt.ylabel('Number of courses per subject')
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.show()
```

# 9. For which levels, Udemy Courses provides the courses

```python
df['level'].value_counts()
```

```python
sns.countplot(x='level', data=df)
plt.xlabel('Level')
plt.ylabel('Number of courses per Level')
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.show()
```

# 10. Display the count of Paid and Free Courses

```python
df['is_paid'].value_counts()
```

# 11. Which course has more lectures (free or paid)?

```python
df.groupby(['is_paid'])['num_lectures'].sum().sort_values(ascending=False)
```

# 12. Which courses have a higher number of subscribers Free or Paid?

```python
df.groupby('is_paid')['num_subscribers'].sum().sort_values(ascending=False)
```

```python
sns.barplot(x='is_paid', y='num_subscribers', data=df)
```

# 13. Which Level has the highest number of subscribers?

```python
df.groupby('level')['num_subscribers'].sum()
```

```python
sns.barplot(x='level', y='num_subscribers', data=df)
plt.xticks(rotation=45)
plt.show()
```

# 14. Find most popular course title

```python
df.groupby('course_title')['num_subscribers'].sum().sort_values(ascending=False).head(1)
```

```python
df[df['num_subscribers']==df['num_subscribers'].max()]['course_title']
```

# 15. Display 10 most popular courses as per number of subscribers

```python
top10 = df.sort_values(by='num_subscribers', ascending=False).head(10)
```

```python
sns.barplot(y='course_title', x='num_subscribers', data=top10)
plt.show()
```

# 16. Find the course which is having the highest number of reviews.

```python
df[df['num_reviews']==df['num_reviews'].max()]['subject']
```

```python
sns.barplot(x='subject', y='num_reviews', data=df)
plt.xticks(rotation=45)
plt.show()
```

# 17. Does Price affect number of reviews?

```python
sns.scatterplot(data=df, x='price', y='num_reviews')
plt.figure(figsize=(15,10))
plt.show()
```

# 18. Find total number of courses related to python

```python
len(df[df['course_title'].str.contains('python', case=False)])
```

# 19. Display 10 most popular python courses as per number of subscribers

```python
python_top10 = df[df['course_title'].str.contains('python', case=False)].sort_values('num_subscribers', ascending=False).head(10)
```

```python
sns.barplot(y='course_title', x='num_subscribers', data=python_top10)
```
