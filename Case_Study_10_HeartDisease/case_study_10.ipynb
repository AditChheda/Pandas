# 9. How many people have heart disease, and how many don't have heart disease?

```python
df['target'].value_counts()
```

```python
sns.countplot(data=df, x='target')
```

# 10. Find count of male and female

```python
df['sex'].value_counts()
```

```python
sns.countplot(data=df, x='sex')
plt.xticks([0, 1], ['female', 'male'])
plt.show()
```

# 11. Find gender distribution according to the target variable

```python
sns.countplot(x='sex', hue='target', data=df)
plt.xticks([0, 1], ['female', 'male'])
plt.legend(labels=['No-Disease', 'Disease'])
plt.show()
```

# 12. Check age distribution

```python
sns.displot(data=df, x='age', bins=20)
plt.show()
```

# 13. Check Chest pain type

```python
sns.countplot(data=df, x='cp')
plt.xticks([0, 1, 2, 3], ['typical angina', 'atypical angina', 'non-angina pain', 'asymptomatic'])
plt.show()
```

# 14. Show the Chest Pain distribution as per target variable

```python
sns.countplot(data=df, x='cp', hue='target')
plt.xticks([0, 1, 2, 3], ['typical angina', 'atypical angina', 'non-angina pain', 'asymptomatic'])
plt.legend(labels=['No-Disease', 'Disease'])
plt.show()
```

# 15. Show Fasting Blood Sugar distribution according to target variable

```python
sns.countplot(data=df, x='fbs', hue='target')
plt.legend(labels=['No-Disease', 'Disease'])
plt.show()
```

# 16. Check Resting Blood Pressure distribution

```python
df['trestbps'].hist()
```

# 17. Compare Resting Blood Pressure as per Sex Column

```python
g = sns.FacetGrid(data=df, hue='sex', aspect=2)
g.map(sns.kdeplot, 'trestbps', fill=True)
plt.legend(labels=['Female', 'Male'])
plt.show()
```
